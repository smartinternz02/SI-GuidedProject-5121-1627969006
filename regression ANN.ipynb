```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer([('one',OneHotEncoder(),[3])],remainder='passthrough')
```

```python
a=ct.fit_transform(a)
```

```python
from sklearn.model_selection import train_test_split 
a_train,a_test,b_train,b_test = train_test_split(a,b,test_size = 0.2,random_state = 0)
```

```python
print(a.shape,"ashape")
print(b.shape,"bshape")
print(a_train.shape,"a_train shape")
print(a_test.shape,"a_test shape")
print(b_train.shape,"b_train shape")
print(b_test.shape,"b_test shape")
```

```python
from sklearn.preprocessing import StandardScaler
d=StandardScaler()
a_train=d.fit_transform(a_train)
a_test=d.fit_transform(a_test)
```

```python
a_train
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense
```

```python
rg= Sequential()
```

```python
a_train.shape
```

```python
rg.add(Dense(units= 6,kernel_initializer='random_uniform',activation='relu'))
```

```python
rg.add(Dense(units=12,kernel_initializer='random_uniform',activation='relu'))
```

```python
rg.add(Dense(units= 1,kernel_initializer='random_uniform',activation='relu'))
```

```python
rg.compile(optimizer='adam',loss='mse', metrics=['mse'])
```

```python
rg.fit(a_train,b_train,epochs=10000)
```

```python
iqi = rg.predict(a_test)
```
